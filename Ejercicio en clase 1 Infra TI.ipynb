## Archivo 2

```python
# Ruta al archivo 2
ruta_2 = '/Users/rodrigoescallon/Documents/Maestria en Ciencia de Datos/Infraestructura y Arquitectura TI/Sesión1/datos/data1.txt'

df2 = pd.read_csv(ruta_2, delimiter=',', quotechar='"', on_bad_lines='skip')

df2.head()
```

```python
# Renombrar columnas y quitar espacios
df2.columns = df2.columns.str.strip().str.replace('"', '')
```

```python
# Limpiar la columna 'City'
df2['City'] = df2['City'].str.strip().str.replace('"', '')
```

```python
df2.shape
```

```python
df2.head()
```

```python
df2.dtypes
```

```python
# Convertir la columna 'Height (inches)' a centímetros y renombrarla a 'Height (cm)' 1 pulgada = 2,54 centímetros
df2['Height (inches)'] = pd.to_numeric(df2['Height (inches)'], errors='coerce') * 2.54
df2.rename(columns={'Height (inches)': 'Height (cm)'}, inplace=True)
```

```python
# Convertir la columna 'Weight (pounds)' a kilogramos y renombrarla a 'Weight (kg)' 1 pound = 0.453592 kg
df2['Weight (pounds)'] = pd.to_numeric(df2['Weight (pounds)'], errors='coerce') * 0.453592
df2.rename(columns={'Weight (pounds)': 'Weight (kg)'}, inplace=True)
```

```python
# Redondear las columnas 'Height (cm)' y 'Weight (kg)' a 1 decimal
df2['Height (cm)'] = df2['Height (cm)'].round(1)
df2['Weight (kg)'] = df2['Weight (kg)'].round(1)
```

```python
df2.head()
```

## Archivo 3

# Unificando los archivos para crear un solo dataset

```python
#Unificando el df1 y df2 ya que tienen las mismas columnas

df= pd.concat([df1,df2])
```

```python
df.head(20)
```

```python
# Establecer la columna 'Name' como índice
#df.set_index('Name', inplace=True)
#df3.set_index('Name', inplace=True)
```

```python
#dataset= df.join(df3)
```

```python
#se realiza un Full Outer Join entre df y df3

dataset = pd.merge(df,df3,on='Name',how='outer',indicator=True)
```

```python
dataset.shape
```

```python
dataset.head(20)
```

```python
#Elimino duplicados del dataset

dataset = dataset.drop_duplicates()
```
